```python
import pandas as pd
from sklearn.neighbors import KDTree
```

```python
#extract data from the csv file
food_trucks_df = pd.read_csv('../dataset/processed/clustered-food-truck-data.csv')
```

```python
input_latitude = 37.7749
input_longitude = -102.4194
target_cluster = 3
```

```python
# Filter data for the specified cluster
cluster_data = food_trucks_df[food_trucks_df['cluster'] == target_cluster]
```

```python
cluster_data.head(10)
```

```python
# Build KD tree for the cluster
kdtree = KDTree(cluster_data[['Latitude', 'Longitude']])
```

```python
# Query the 5 nearest neighbors to the input coordinates
_, indices = kdtree.query([[input_latitude, input_longitude]], k=5)
```

```python
# Retrieve the 5 closest food trucks
closest_food_trucks = cluster_data.iloc[indices[0]]

print("5 Closest Food Trucks:")
#print(closest_food_trucks)

print(closest_food_trucks[['Latitude', 'Longitude']])
```
